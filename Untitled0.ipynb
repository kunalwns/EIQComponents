```python
import pandas as pd
import matplotlib.pyplot as plt

# Read Excel data
data = pd.read_excel('/content/sample_data/Postbind_report.xlsx')

# Assuming your Excel has multiple sets of data in different columns
x1 = data['Post Bind Task ID']
y1 = data['PolicyReference']

x2 = data['BrokerNumber']
y2 = data['Broker_Name']

output_width_inches = 15  # You may need to adjust this value based on your requirement

# Calculate the height of the figure based on the aspect ratio of your plot
aspect_ratio = 6/3  # You may need to adjust this value based on your requirement
output_height_inches = output_width_inches / aspect_ratio

# Generate plots
plt.figure(figsize=(output_width_inches, output_height_inches))

# Plot first graph
plt.subplot(1, 2, 1)
plt.plot(x1, y1)
plt.xlabel('Task ID')
plt.ylabel('Policy Reference')
plt.title('Policy Reference detail')
plt.grid(True)

# Plot second graph
plt.subplot(1, 2, 2)
plt.plot(x2, y2)
plt.xlabel('Broker Number')
plt.ylabel('Broker Name')
plt.title('Broker Name Details')
plt.grid(True)

# Save the plots as images
plt.savefig('output_graph.png')

# Save the data to an Excel file
with pd.ExcelWriter('output.xlsx') as writer:
    data.to_excel(writer, sheet_name='Data')

# Display the plots
plt.show()
```
